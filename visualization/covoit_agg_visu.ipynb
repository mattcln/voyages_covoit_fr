```python
import pandas as pd
import os
import plotly.express as px
```

```python
in_relative_path = "../outputs/"
in_absolute_path = os.path.abspath(os.path.join(os.getcwd(), in_relative_path))

df = pd.read_csv(f"{in_absolute_path}/reduced_concat_covoit.csv")
df
```

# En prenant compte des trajets hors-plateforme  
D'aprÃ¨s [Comprendre le covoiturage quotidien en France](https://observatoire.covoiturage.gouv.fr/observatoire/comprendre-covoiturage-quotidien/), 96% des trajets covoiturÃ©s le sont de maniÃ¨re informelle.
Il nous reste donc Ã  multiplier les rÃ©sultats par 25 pour obtenir les chiffres approximatifs du covoiturage en France.

```python
df["nb_passagers"] = df["nb_passagers"] * 25
df["nb_trajets"] = df["nb_trajets"] * 25
df["total_distance"] = df["total_distance"] * 25
```

```python
fig = px.line(df, x="journey_start_date", y="nb_passagers", title='Evolution du nombre de trajets de covoiturage quotidien (plateforme + hors-plateforme)')
fig.show()
```

Chaque semaine, on remarque un pic le mardi et une baisse significative (~66%) le week-end.

```python
fig = px.line(df, x="journey_start_date", y="total_distance", title='Evolution de la distance total en covoiturage au quotidien (plateforme + hors-plateforme)')
fig.show()
```
